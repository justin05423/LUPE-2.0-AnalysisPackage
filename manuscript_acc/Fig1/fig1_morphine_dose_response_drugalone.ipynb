```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import sys

if not os.path.join(os.path.abspath(''), '../') in sys.path:
    sys.path.append(os.path.join(os.path.abspath(''), '../'))

from utils.classification import load_behaviors
from utils.meta import *
```

### load predicted behaviors

```python
behaviors = load_behaviors('../processed_dataset/behaviors.pkl')
```

```python
### some params
# bin time 1 minute, change the last  number to 2 for 2 minute bins
num_min = 1
time_bin_size = 60*60*num_min
# take random file to compute total number of bins
n_bins = len(behaviors['Male']['Group1']['Condition1']['file0'])/time_bin_size

# 0mg/kg morphine
selected_group_exp0 = 'Group1'
selected_condition_exp0 = 'Condition1'

# 0.5mg/kg morphine
selected_group_exp1 = 'Group2'
selected_condition_exp1 = 'Condition1'

# 1mg/kg morphine
selected_group_exp2 = 'Group3'
selected_condition_exp2 = 'Condition1'

# 5mg/kg morphine
selected_group_exp4 = 'Group4'
selected_condition_exp4 = 'Condition1'

# 10mg/kg morphine
selected_group_exp5 = 'Group5'
selected_condition_exp5 = 'Condition1'

annotations = ['0mg/kg morphine', '0.5mg/kg morphine', '1mg/kg morphine', '5mg/kg morphine', '10mg/kg morphine']
```

```python
# bin behaviors per params above
selected_groups = [selected_group_exp0, selected_group_exp1, selected_group_exp2, selected_group_exp4, selected_group_exp5]
selected_conditions = [selected_condition_exp0, selected_condition_exp1, selected_condition_exp2, selected_condition_exp4, selected_condition_exp5]
rows = 1
cols = len(selected_groups)
count = 0
fig, ax = plt.subplots(rows, cols, figsize=(12, 2.5), sharey=True)
y_to_save = {key: [] for key in annotations}
for selected_group, selected_condition in zip(selected_groups, selected_conditions):
    
    behavior_ratios_across_sex = {key: np.NAN 
                                  for key in sexes}
    for sex in sexes:
        behavior_ratios_files = {key: np.NAN 
                                 for key in range(len(behaviors[sex][selected_group][selected_condition]))}
        
        for f in range(len(behaviors[sex][selected_group][selected_condition])):
            binned_behavs = []
            for bin_n in range(int(n_bins)):
                # get counts paired with unique value
                behavior_ratios = {key: 0 for key in range(len(behavior_names))}
                values, counts = np.unique(
                    behaviors[sex][selected_group][selected_condition][f'file{f}'][
                    time_bin_size*bin_n:time_bin_size*(bin_n+1)], 
                    return_counts=True)
                # normalize to sum = 1
                for i, value in enumerate(values):
                    behavior_ratios[value] = counts[i]/sum(counts)
                binned_behavs.append(behavior_ratios)
            # store relative to file number
            behavior_ratios_files[f] = binned_behavs
        # store relative to sex
        behavior_ratios_across_sex[sex] = behavior_ratios_files
    # plotting below, compute mean and sem across all files (regardless of sex)
    y_to_save[annotations[count]] = {key: [] for key in behavior_names}
    for b in range(len(behavior_names)):
        y_files = []
        for sex in sexes:
            for f in range(len(behavior_ratios_across_sex[sex])):
                y_files.append(np.hstack([behavior_ratios_across_sex[sex][f][bin][b] 
                                          for bin in range(len(behavior_ratios_across_sex[sex][f]))])
                               )
        y = (np.mean(y_files, axis=0))
        y_to_save[annotations[count]][behavior_names[b]] = y.copy()
        x = np.arange(int(n_bins))
        y_sem = (np.std(y_files, axis=0)/np.sqrt(len(behavior_ratios_files)))
        # plot shadederrorbar
        ax[count].plot(x, y, color=behavior_colors[b])
        ax[count].fill_between(x, y-y_sem, y+y_sem, color=behavior_colors[b], alpha=0.2)
    ax[count].set_title(annotations[count])
    ax[count].spines[['top', 'right']].set_visible(False)
    ax[count].set_xlabel(f'Time bin = {num_min}min')
    ax[count].set_xticks(np.arange(0, n_bins, 10))
    
    if count == 0:
        ax[count].set_ylabel('Probability')
    count += 1
```

```python
df = pd.DataFrame(y_to_save).T
df.to_csv('../processed_dataset/fig1_morphine_dose_response_drugalone/mean_ratio.csv')
```
