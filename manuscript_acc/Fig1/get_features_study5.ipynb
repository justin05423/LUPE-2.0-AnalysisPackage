### get features from pose

```python
if 'binned_features_study5.pkl' not in os.listdir('../processed_dataset/'):
    features = {key: [] for key in drugs_study5}
    for drug in notebook.tqdm(drugs_study5):
        features[drug] = {key: [] for key in groups_study5}
        for group in notebook.tqdm(groups_study5):
            features[drug][group] = {key: [] for key in conditions_study5}
            for condition in notebook.tqdm(conditions_study5):
                features[drug][group][condition] = {key: [] for key in [f'file{i}' 
                                                   for i in range(len(data[drug][group][condition]))]}

                for i in range(len(data[drug][group][condition])):
                    features[drug][group][condition][f'file{i}'] = (
                        feature_extraction([data[drug][group][condition][f'file{i}']], 1, framerate=60))
else:
    features = load_features('../processed_dataset/binned_features_study5.pkl')
```

```python
if 'binned_features_study5.pkl' not in os.listdir('../processed_dataset/'):
    with open('../processed_dataset/binned_features_study5.pkl', 'wb') as f:
        pickle.dump(features, f)
else:
    print('Binned features exists!')
```
