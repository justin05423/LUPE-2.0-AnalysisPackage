### get features from pose

```python
if 'binned_features_sni.pkl' not in os.listdir('../processed_dataset/'):
    features = {key: [] for key in groups_sni}
    for group in notebook.tqdm(groups_sni):
        features[group] = {key: [] for key in conditions}
        for condition in notebook.tqdm(conditions):
            features[group][condition] = {key: [] for key in [f'file{i}' 
                                               for i in range(len(data[group][condition]))]}

            for i in range(len(data[group][condition])):
                features[group][condition][f'file{i}'] = (
                    feature_extraction([data[group][condition][f'file{i}']], 1, framerate=60))
else:
    features = load_features('../processed_dataset/binned_features_sni.pkl')
```

```python
if 'binned_features_sni.pkl' not in os.listdir('../processed_dataset/'):
    with open('../processed_dataset/binned_features_sni.pkl', 'wb') as f:
        pickle.dump(features, f)
else:
    print('Binned features exists!')
```
