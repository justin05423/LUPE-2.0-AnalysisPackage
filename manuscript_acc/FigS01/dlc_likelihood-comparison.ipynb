```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load CSV files into DataFrame objects
df1 = pd.read_csv('F_LUPE-hab_likelihood.csv', header=None)
df2 = pd.read_csv('M_LUPE-hab_likelihood.csv', header=None)

# Define the bin size
bin_size = 300

# Bin the data and calculate the average likelihood for each bin
binned_df1 = df1.groupby(np.arange(len(df1)) // bin_size).mean()
binned_df2 = df2.groupby(np.arange(len(df2)) // bin_size).mean()

# Plot the line graphs for each column comparison
num_columns = min(len(binned_df1.columns), len(binned_df2.columns))
fig, axes = plt.subplots(nrows=num_columns, ncols=1, figsize=(10, 5*num_columns))

for i, (col1, col2) in enumerate(zip(binned_df1.columns, binned_df2.columns)):
    ax = axes[i] if num_columns > 1 else axes
    ax.plot(binned_df1.index, binned_df1[col1], label='Dataset 1')
    ax.plot(binned_df2.index, binned_df2[col2], label='Dataset 2')
    ax.set_title(f'Comparison of {col1} and {col2}')
    ax.set_xlabel('Bin Index')
    ax.set_ylabel('Likelihood Value')
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV files into DataFrame objects
df1 = pd.read_csv('F_LUPE-hab_likelihood.csv', header=None)
df2 = pd.read_csv('M_LUPE-hab_likelihood.csv', header=None)

# Define the keypoints
keypoints = ["nose", "mouth", "l_forepaw", "l_forepaw_digit", "r_forepaw", "r_forepaw_digit",
             "l_hindpaw", "l_hindpaw_digit1", "l_hindpaw_digit2", "l_hindpaw_digit3",
             "l_hindpaw_digit4", "l_hindpaw_digit5", "r_hindpaw", "r_hindpaw_digit1",
             "r_hindpaw_digit2", "r_hindpaw_digit3", "r_hindpaw_digit4", "r_hindpaw_digit5",
             "genitalia", "tail_base"]

# Calculate the average likelihood for each column in both datasets
avg_likelihood_df1 = df1.mean()
avg_likelihood_df2 = df2.mean()

# Create a scatter plot to compare the average likelihood of each column between the two files
plt.figure(figsize=(10, 6))
plt.scatter(range(len(avg_likelihood_df1)), avg_likelihood_df1, label='Male Pose Probability')
plt.scatter(range(len(avg_likelihood_df2)), avg_likelihood_df2, label='Female Pose Probability')
plt.title('LUPE Habituation - Average Likelihood of Body-Points Single Male and Female Mice')
plt.xlabel('Body-Points')
plt.ylabel('Average Likelihood')
plt.xticks(range(len(avg_likelihood_df1)), keypoints, rotation=45, ha='right')  # Set x-axis ticks to correspond to keypoints
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.savefig('scatter_plot.svg', format='svg')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load CSV files into DataFrame objects
df1 = pd.read_csv('F_LUPE-hab_likelihood.csv', header=None)
df2 = pd.read_csv('M_LUPE-hab_likelihood.csv', header=None)

# Define the keypoints
keypoints = ["nose", "mouth", "l_forepaw", "l_forepaw_digit", "r_forepaw", "r_forepaw_digit",
             "l_hindpaw", "l_hindpaw_digit1", "l_hindpaw_digit2", "l_hindpaw_digit3",
             "l_hindpaw_digit4", "l_hindpaw_digit5", "r_hindpaw", "r_hindpaw_digit1",
             "r_hindpaw_digit2", "r_hindpaw_digit3", "r_hindpaw_digit4", "r_hindpaw_digit5",
             "genitalia", "tail_base"]

# Calculate the average likelihood for each column in both datasets
avg_likelihood_df1 = df1.mean()
avg_likelihood_df2 = df2.mean()

# Calculate the absolute difference in average likelihood between the two datasets for each body part
diff_likelihood = np.abs(avg_likelihood_df1 - avg_likelihood_df2)

# Create a scatter plot to compare the average likelihood of each column between the two files
plt.figure(figsize=(10, 6))
scatter = plt.scatter(range(len(avg_likelihood_df1)), avg_likelihood_df1, label='Male Pose Probability',
                      c=diff_likelihood, cmap='coolwarm', s=100)
plt.scatter(range(len(avg_likelihood_df2)), avg_likelihood_df2, label='Female Pose Probability', c=diff_likelihood,
            cmap='coolwarm', s=100, marker='^')
plt.colorbar(scatter, label='Absolute Difference in Likelihood')
plt.title('LUPE Habituation - Average Likelihood of Body-Points Single Male and Female Mice')
plt.xlabel('Body-Points')
plt.ylabel('Average Likelihood')
plt.xticks(range(len(avg_likelihood_df1)), keypoints, rotation=45, ha='right')  # Set x-axis ticks to correspond to keypoints
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.savefig('scatter_plot_difference.svg', format='svg')


plt.show()


```
